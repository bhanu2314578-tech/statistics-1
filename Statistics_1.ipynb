**Question 1: What is a random variable in probability theory?**

Answer:- A random variable is a numerical quantity whose value depends on the outcome of a random event or experiment. It assigns a real number to each possible outcome in a sample space and is used to quantify uncertainty in probability and statistics.

**Question 2: What are the types of random variables?**

Answer:- There are two main types of random variables:-

**Discrete Random Variable:** Takes Specific, countable values(e.g; number of heads in 5 coin tosses).
**Continuous Random Variable:** Takes infinite number of values within a range (e.g; height, weight , temperature)


**Question 3: Explain the difference between discrete and continuous distributions.**

Answer:-
**Discrete Distribution:** Represents probabilities for discrete random variables. Each outcome has a specific probability and the total probability equals 1 (e.g; Binomial distribution) .


 **Continuous Distribution:** Represents probabilities for continuous variables using a Probability Density Function (PDF). Probabilities are calculated over intervals (e.g; Normal Distribution)


**Question 4: What is a binomial distribution, and how is it used in probability?**

Answer: A binomial distribution describes the number of successes in a fixed number of independent Bernoulli trials , each with the same probability of success p.

Formula:   P(X=k) = (n k) p^k (1-p)^(n-k)

It is used in scenarios like quality testing, survey sampling and reliability analysis Where outcomes are binary (success/failure).

**Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

Answer: The Central Limit Theorem states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population's original distribution. It is critical because it allows statisticians to apply normal-based methods like confidence intervals and hypothesis tests even for non-normal populations.

**Question 7: What is the significance of confidence intervals in statistical analysis?**

Answer: A confidence interval gives a range of values likely to contain the true population parameter with a certain level of confidence. It gives a measure of estimate precision  narrower intervals indicate more accurate estimate and is essential for decision-making under uncertainty.

**Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.**

```python
#Answer:

import numpy as np
import matplotlib.pyplot as plt

# Generate random numbers
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean_value = np.mean(data)
std_value = np.std(data)

print(f"Mean: {mean_value:.2f}")
print(f"Standard Deviation: {std_value:.2f}")

# Plot histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Normal Distribution (mean=50, std=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

**Question 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend. daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]**

**● Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.**

 **● Write the Python code to compute the mean sales and its confidence interval.
 (Include your Python code and output in the code box below.)**

 Answer: Answer: The Central Limit Theorem allows us to approx the sampling distribution of the mean as normal when the sample size is large enough.
To estimate the 95% confidence interval, we use:

CI = x‾ - z @/ 2 *  s/ ✓n

Where x is the sample mean, s is the sample standard deviation and z @/2= 1.96

```python
#Python Code:-

import numpy as np
import scipy.stats as stats

# Daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Compute statistics
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)
n = len(daily_sales)

# 95% confidence interval
confidence_level = 0.95
z_value = stats.norm.ppf(1 - (1 - confidence_level)/2)
margin_of_error = z_value * (std_sales / np.sqrt(n))
ci_lower = mean_sales - margin_of_error
ci_upper = mean_sales + margin_of_error

print(f"Mean Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```
